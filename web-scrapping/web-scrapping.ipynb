```python
final_df = pd.DataFrame()
for j in range(1,50):
    
    url = 'https://www.ambitionbox.com/list-of-companies?page={}'.format(j)
    headers={'User-Agent':'Mozilla/5.0(Windows NT 6.3;Win64; x64) AppleWebkit/537.36 (KHTML,like Gecko) Chrome/80.0.3987/162 Safari/537.36'}
    webpage = requests.get(url,headers=headers).text

    soup = BeautifulSoup(webpage,"html.parser") 

    company = soup.find_all('div',attrs={"class": "company-content-wrapper"})
    
    name = []
    rating = []
    ctype = []
    location=[]
    old = []
    reviews = []
    emp = []

    for i in company:
        
        ### print(i.find('h2').text.strip())
        name.append(i.find('h2').text.strip())
        rating.append(i.find('p',attrs={'class':'rating'}).text.strip())
        try:
            ctype.append(i.find_all('p',attrs={'class':'infoEntity'})[0].text.strip())
        except:
            ctype.append(np.nan)
        try:
            location.append(i.find_all('p',attrs={'class':'infoEntity'})[1].text.strip())
        except:
            location.append(np.nan)
        try:
            old.append(i.find_all('p',attrs={'class':'infoEntity'})[2].text.strip())
        except:
            old.append(np.nan)
        try:
            emp.append(i.find_all('p',attrs={'class':'infoEntity'})[3].text.strip())
        except:
            emp.append(np.nan)

        reviews.append(i.find('a',attrs={'class':'review-count'}).text.strip())

    dict={'name':name,'rating':rating,'company_type':ctype,'location':location,'old':old,'emp':emp,'reviews':reviews}
    df = pd.DataFrame(dict)
    final_df = final_df.append(df,ignore_index=True)
    final_df
```
