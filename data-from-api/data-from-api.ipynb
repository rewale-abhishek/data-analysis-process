### Getting Data from API as CSV File. TMDB Api

```python
### importing required packages.
import requests
import pandas as pd
```

```python
### hitting the required url in request.get() method.
response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=<paste-your-key>&language=en-US&page=1').json()
response
```

```python
### looping through the list.

for i in response['results']:
    print(i['title'],i['vote_average'])
```

```python
### fetching data from only from page no. 1.

title = []
original_language = []
popularity = []
vote_count = []
vote_average = []
release_date = []

for i in response['results']:
    title.append(i['title'])
    original_language.append(i['original_language'])
    popularity.append(i['popularity'])
    vote_count.append(i['vote_count'])
    vote_average.append(i['vote_average'])
    release_date.append(i['release_date'])
    
dict = {'title':title,
        'original_language':original_language,
        'popularity':popularity,
        'vote_count':vote_count,
        'vote_average':vote_average,
        'release_date':release_date}

df = pd.DataFrame(dict)
```

```python
### getting the data from 

final_df = pd.DataFrame()

for j in range(1,50):
    
    response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=1652441d2fafb12010d4cdcb2d7628c1&language=en-US&page={}'.format(j)).json()
        
    title = []
    original_language = []
    popularity = []
    vote_count = []
    vote_average = []
    release_date = []

    for i in response['results']:
        title.append(i['title'])
        original_language.append(i['original_language'])
        popularity.append(i['popularity'])
        vote_count.append(i['vote_count'])
        vote_average.append(i['vote_average'])
        

    dict = {'title':title,
            'original_language':original_language,
            'popularity':popularity,
            'vote_count':vote_count,
            'vote_average':vote_average}
            

    df = pd.DataFrame(dict)


    final_df = final_df.append(df,ignore_index=True)
```
